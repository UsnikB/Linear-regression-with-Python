# Importing all ther nescessery Modules

```python
import numpy as np # Numpy used for linear algebra
import pandas as pd # Pandas is used for 
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt
from sklearn.metrics import mean_squared_error
from sklearn.metrics import mean_squared_error, r2_score, mean_absolute_error
```

# Importing and describing the data

```python
data=pd.read_csv('../data/NFLX.csv')
data.head()
```

# Plotting the data before applying leniar regression

```python
plt.figure(figsize=(10, 5))
data['Close'].plot()
```

# Cleaning the data

```python
# data = data.reset_index() # Resetting the index as sometime the index might have invalid sequence
x = np.array(data.index).reshape(-1, 1) 
y = data['Close'] # getting the relavent data column form the the data table
data= data.dropna() # removing all the rows with empty/no data
```

# Splitting the data in 80 20%

```python
x_train, x_test, y_train, y_test = train_test_split(x,y, test_size=.2)
```

# Applying linear regression from siket learn

```python
linreg = LinearRegression().fit(x_train, y_train)
y_test = linreg.predict(x_test)
print(linreg.score(x_train, y_train))
```

```python
print("Model Coefficients:", linreg.coef_)
print("Mean Absolute Error:", mean_absolute_error(y_train, x_train))
print("Coefficient of Determination:", r2_score(x_test, y_test))
```

# Plotting the result

```python
plt.figure(figsize=(15,5))
plt.plot(data['Close'])
plt.plot(x_test, y_test)
```
